 ### Demographic Analysis

**Age distribution of respondents**


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution of Respondents', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

**Gender distribution of respondents**

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Gender', hue='Gender', palette="Set2", legend=False)
plt.title('Gender Distribution of Respondents', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

## Mental Health Condition Analysis

**Prevalence of Mental Health Conditions:** Analyze how many respondents report mental health conditions.


```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Mental_Health_Condition', hue='Mental_Health_Condition', palette="Set2", legend=False)
plt.title('Distribution of Mental Health Conditions', fontsize=16)
plt.xlabel('Mental Health Condition', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

**Severity Distribution:** Visualize the severity of mental health conditions reported.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Severity', hue='Severity', palette="Set2", legend=False)
plt.title('Severity of Mental Health Conditions', fontsize=16)
plt.xlabel('Severity', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

## Consultation History

**Consultation History Distribution :** Examine how many respondents have sought consultation for mental health issues.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Consultation_History', hue='Consultation_History', palette="Set2", legend=False)
plt.title('Consultation History for Mental Health', fontsize=16)
plt.xlabel('Consultation History', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

##  Lifestyle Factors

**Stress Level Analysis:** Analyze the stress levels reported by respondents.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Stress_Level', hue='Stress_Level', palette="Set2", legend=False)
plt.title('Stress Level of Respondents', fontsize=16)
plt.xlabel('Stress Level', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

**Sleep Hours Analysis:** Visualize the distribution of sleep hours among respondents.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sleep_Hours'], bins=15, kde=True,color="skyblue" )
plt.title('Sleep Hours of Respondents', fontsize=16)
plt.xlabel('Sleep Hours', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

**Work Hours Analysis:** Examine the number of work hours reported.

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.histplot(df['Work_Hours'], bins=15, kde=True, color="skyblue")
plt.title('Work Hours of Respondents', fontsize=16)
plt.xlabel('Work Hours', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

**Physical Activity Hours Analysis:** Analyze physical activity hours reported.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Physical_Activity_Hours'], bins=10, kde=True)
plt.title('Physical Activity Hours of Respondents', fontsize=16)
plt.xlabel('Physical Activity Hours', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```
